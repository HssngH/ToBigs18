# Data Cleansing

```python
# 변수 치환
## dictionary 사용하여 key값을 기준으로 찾아 치환될 수 있도록 작성함.
Education_dic = { 1 : 'Below College', 2 : 'College', 3 : 'Bachelor', 4 : 'Master', 5 : 'Doctor'}
EnvironmentSatisfaction_dic = { 1 : 'Low', 2 : 'Medium', 3 : 'High', 4 : 'Very High'}
JobInvolvement_dic = { 1 : 'Low', 2 : 'Medium', 3 : 'High', 4 : 'Very High'}
JobSatisfaction_dic = { 1 : 'Low', 2 : 'Medium', 3 : 'High', 4 : 'Very High'}
PerformanceRating_dic = { 1 : 'Low', 2 : 'Good', 3 : 'Excellent', 4 :'Outstanding'}
RelationshipSatisfaction_dic = { 1 : 'Low', 2 : 'Medium', 3 :'High', 4 : 'Very High'}
WorkLifeBalance_dic = { 1 : 'Bad', 2 : 'Good', 3 : 'Better', 4 : 'Best'}

df.replace( {'Education' : Education_dic ,  'EnvironmentSatisfaction' : EnvironmentSatisfaction_dic, 'JobInvolvement' : JobInvolvement_dic,
             'JobSatisfaction' : JobSatisfaction_dic, 'PerformanceRating' : PerformanceRating_dic, 
             'RelationshipSatisfaction' : RelationshipSatisfaction_dic, 'WorkLifeBalance' : WorkLifeBalance_dic}, inplace=True)
```

```python
# 변수별 값 분포 확인하기
## 치환이 잘 되었는지도 확인하기
for col in df.columns:
    print('변수',col,'의 분포는 :')
    print(df[col].value_counts())
    print('-----------------------')
```

```python
# 모든 값이 동일한 변수 삭제
df.drop( ['EmployeeCount', 'Over18', 'StandardHours'], axis=1, inplace=True)

# 모든 값이 다른 변수 삭제
df.drop( ['EmployeeNumber'], axis=1, inplace=True)
```

```python
# 변수를 데이터 타입에 따라 분리해서 리스트 생성해주기 (그래프 그리기 위해서)
int_list = []
obj_list = []

for col in df.columns:
    if df[col].dtype == 'int64':
        int_list.append(col)
    else:
        obj_list.append(col)

print('DataType : int', int_list)
print('DataType : obj', obj_list)
```

```python
# 직업 만족도는 업무의 참여도와 근속 연수에 유의한 영향을 끼친다.
## x값이 근속연수, y값이 업무 참여도
satis_level = df['JobSatisfaction'].unique().tolist()

for satis in satis_level:
    x = df['JobInvolvement'][df['JobSatisfaction']==satis].value_counts().index.tolist()
    y = df['JobInvolvement'][df['JobSatisfaction']==satis].value_counts().tolist()
    colors=['orange','red','blue', 'green']
    plt.title('JobSatisfaction :  '+satis)
    plt.barh(x, y, color=colors)
    plt.ylabel('YearsAtCompany', fontsize=10)
    plt.xlabel('JobInvolvement', fontsize=10)
    plt.xlim(0, 300)
    plt.show()

## 직업 만족도가 Very High, High인 경우에는 근속연수도 높고 업무의 참여도도 높다.
## 하지만 직업 만족도가 Medium, Low인 경우에는 근속연수도 낮고 업무의 참여도도 낮다.
```

```python
# 파생변수
## 업무 성과도에 따른 급여의 증가분 백분율로 급여 증가율 단계 나누기

# 업무 성과도 Excellent의 범위
ex_q1=df['PercentSalaryHike'][df['PerformanceRating']=='Excellent'].quantile(0.25)
ex_q3=df['PercentSalaryHike'][df['PerformanceRating']=='Excellent'].quantile(0.75)
ex_q2=df['PercentSalaryHike'][df['PerformanceRating']=='Excellent'].quantile(0.5)
ex_min=df['PercentSalaryHike'][df['PerformanceRating']=='Excellent'].min()
ex_max=df['PercentSalaryHike'][df['PerformanceRating']=='Excellent'].max()

# 업무 성과도 Outstanding의 범위
out_q1=df['PercentSalaryHike'][df['PerformanceRating']=='Outstanding'].quantile(0.25)
out_q3=df['PercentSalaryHike'][df['PerformanceRating']=='Outstanding'].quantile(0.75)
out_q2=df['PercentSalaryHike'][df['PerformanceRating']=='Outstanding'].quantile(0.5)
out_min=df['PercentSalaryHike'][df['PerformanceRating']=='Outstanding'].min()
out_max=df['PercentSalaryHike'][df['PerformanceRating']=='Outstanding'].max()


# 업무 성과도에 따른 임금 증가 여부 정해주는 함수 생성
def PerformanceRating(x):
    x = int(x)
    if x >=out_q3 or out_min>=x>ex_q3 :
        x = 'Super Increase'
    elif  ex_q3>=x>=ex_q2 or out_q3>=x>=out_q2:
        x = 'Increase'
    else:
        x = 'No Increase'
    return x
```

```python
df['SalaryIncrease'] = df['PercentSalaryHike'].apply(PerformanceRating)
```

```python
# 임금 상승 비중
labels = df['SalaryIncrease'].value_counts().index.tolist()
ratio = df['SalaryIncrease'].value_counts()

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, shadow=True)
plt.show()
```
